# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
group_by_neighbourhood_group = df.groupby('neighbourhood_group')
group_by_neighbourhood_group.size()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
group_by_neighbourhood_group['price'].mean()

```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
group_by_neighbourhood_group['price'].agg(['max', 'min'])

    
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
group_by_neighbourhood_group_and_room_type= df.groupby(['neighbourhood_group', 'room_type'])
group_by_neighbourhood_group_and_room_type['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
group_by_neighbourhood_group_and_room_type['price'].agg(['count', 'min', 'max', 'mean', 'median', 'std'])
```

```python
# #  1. Who was won Album of the Year in 2016?
# test = grammys[grammys['year']==2016]
# test = test[test['category'] == 'Album of the Year']
# # #grammys[grammys.isin(['2016','Album of the Year'])]
#test = (grammys['year']== '2016.0') & (grammys['category']== 'Album of the Year') & (grammys['winner']== True)  
a = grammys.where((grammys['year']== 2016) & (grammys['category']== 'Album of the Year') & (grammys['winner']== True),inplace = False).dropna()['nominee']
a
```

```python
# 2. Who won Best Rap Album in 2009?
b = grammys.where((grammys['year']== 2009) & (grammys['category']== 'Best Rap Album') & (grammys['winner']== True),inplace = False).dropna()['nominee']
b
```

```python
# 3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?
kl = grammys[grammys.apply(lambda row: row.astype(str).str.contains('Kendrick Lamar').any(), axis=1)]
print('Kendrick Lamar, in one capacity or another, was nominated for ', len(kl), 'awards and won', kl['winner'].sum(), 'of them')
```
