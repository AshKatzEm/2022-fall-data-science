# Using Stocks.csv
* Tell me a story about the Google stock for the year. 
* Any story you'd like to tell. 


```python
stock = pd.read_csv('data/Stocks.csv')
stock
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data=stock,
             x="Date",
             y="Volume",
             ax=ax      
            )
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data=stock,
             x="Date",
             y="Close",
             ax=ax      
            )
```

```python
stock['Month'] = 0
for i in stock.index:
    stock['Month'][i] = datetime.strptime(stock["Date"][i], '%Y-%m-%d').strftime('%m-%Y')
```

```python
stockMonth = stock.groupby(by='Month',
                  axis=0,
                  level=None,
                  as_index=True,
                  sort=True,
                  observed=False,
                  dropna=True
                 )
stockMonthMedian = stockMonth.median()
stockMonthMedian
#stockMonthMedian = stockMonthMedian[['Close', 'Volume']]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(data= stockMonthMedian,
                x='Month',
                y='Close',
                hue='Volume',
                size='Volume',
                style=None,
                palette='hls',
                hue_order=None,
                hue_norm=None,
                sizes= (100,1000),
                size_order=None,
                size_norm=None,
                markers=True,
                style_order=None,
                legend= 'full',
                ax=ax      
               )
```

```python
print('As you can see, the stock has the highest trade volume when the price is low')
print('People when buy and selling less and less as the price went up, peaking in June')
print('But then something happened. Maybe the stock price is not going down again')
print('Maybe $2750 is a good price because its only going to go up from here')
print('As you can see, the trade volume begins to clime again in July, despite the price increases')
print('Maybe there was a good earnings report')


```
